```python
raw_df['Location'] = raw_df['Location'].astype(str).str.split(',').str[1].fillna(raw_df['Location']).str.strip()
print(raw_df['Location'].head())
```

```python
central_list = ['West Mambalam','Parrys','Anna Salai','Saidapet','Arumbakkam','RK Salai (Cathedral Road)','T.Nagar','Egmore','Nandanam','Shenoy Nagar','Thousand Lights','Abhiramapuram','Choolaimedu','West Chennai Mambalam','RA Puram','Chetpet','Kilpauk','Purasavakkam','Aminijikarai','T. Nagar','Nungambakkam','Teynampet','Alwarpet', 'Boat Club', 'Poes Garden', 'Gopalapuram','R.A. Puram','Mylapore', 'Triplicane', 'Chepauk', 'Royapettah','Kodambakkam','Vadapalani']
south_list = ['Vettuvankeni','Oragadam','Old Mahabalipuram Road','Perumbakkam','Kelambakkam','Santhome','Okkiyampet','Guduvancheri','Mamallapuram','Coast Road (ECR)','OMR','Kandanchavadi','Navallur','Nanganallur','Padur', 'Kottivakkam','Kovalam','Vandalur','Pallikaranai','Karapakkam','Muttukadu','Palavakkam','Ekkaduthangal','Adambakkam','Akkarai','St. Thomas Mount','Egatoor','MRC Nagar','GST Road', 'Selaiyur','Semmancheri','Potheri','Thuraipakkam','Porur','K.K. Nagar','Ashok Nagar','Adyar', 'Besant Nagar','Kotturpuram', 'Mandaveli','Alandur','East Coast Road (ECR)','ECR','Kanathur','Mahabalipuram','Thiruvanmiyur', 'Taramani', 'Perungudi', 'Sholinganallur','Neelangarai','Injambakkam','Guindy', 'Tambaram', 'Chromepet', 'Pallavaram','Meenambakkam','Velachery','Medavakkam','Madipakkam','Chitlapakkam','Puzhuthivakkam',' Kanchipuram','Kanchipuram District']
west_list = ['Virugambakkam','Sriperumbudur','Thiruvallur','Koyambedu','Poonamalle','Saligramam','Ramapuram','Anna Nagar', 'Valasaravakkam', 'Mogappair', 'Maduravoyal', 'Iyyappanthangal','Kolathur','Ambattur', 'Padi','Avadi','Thiruninravur']
north_list = ['Anna Nagar West''Anna Nagar East','Vepery','Tiruvottiyur','Redhills','Saidapet','Chengalpattu','George Town','Sowcarpet','Thiruvottiyur','Manali', 'Madhavaram','Tondiarpet','Perambur','Washermenpet','Royapuram','Kanchipuram', 'Old Washermenpet','New Washermenpet','Vyasarpadi','Korukkupet','Erukanchery, Kodungaiyur','Sembiam','Jamalia','Red Hills','Pattalam', 'Otteri', 'Kosapet', 'Pulianthope', 'Park Town']
def categorize_chennai(location):
    if location in central_list:
        return 'Central Chennai'
    elif location in south_list:
        return 'South Chennai'
    elif location in west_list:
        return 'West Chennai'
    elif location in north_list:
        return 'North Chennai'
    else:
        return 'Other/Unknown'
```

```python
delivery_count = zomato_df['Features'].str.contains('Home Delivery', case=False, na=False).sum()
print(f"Total restaurants offering Home Delivery: {delivery_count}")
delivery_count = zomato_df['Features'].str.contains('Indoor Seating', case=False, na=False).sum()
print(f"Total restaurants offering Indoor Seating: {delivery_count}")
delivery_count = zomato_df['Features'].str.contains('Table booking recommended', case=False, na=False).sum()
print(f"Total restaurants offering Table booking recommended: {delivery_count}")
delivery_count = zomato_df['Features'].str.contains('Vegetarian Only', case=False, na=False).sum()
print(f"Total restaurants offering Vegetarian Only: {delivery_count}")
delivery_count = zomato_df['Features'].str.contains('Full Bar Available', case=False, na=False).sum()
print(f"Total restaurants offering Full Bar Available: {delivery_count}")
delivery_count = zomato_df['Features'].str.contains('Valet Parking Available', case=False, na=False).sum()
print(f"Total restaurants offering Valet Parking Available: {delivery_count}")
```

```python
features_to_check = ['Home Delivery', 'Indoor Seating', 'Table booking recommended', 'Vegetarian Only', 'Full Bar Available', 'Valet Parking Available']
results = {}
for feature in features_to_check:
    count = zomato_df['Features'].str.contains(feature, case=False, na=False).sum()
    results[feature] = count
feature_type_df = pd.DataFrame(list(results.items()), columns=['Feature Type', 'Restaurant Count'])
print(feature_type_df)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Restaurant Count', y='Feature Type', data=feature_type_df)
plt.title('COUNT OF RESTAURANTS BASED ON FEATURE AVAILABLE')
plt.xlabel('Number of Restaurants')
plt.ylabel('Feature Type')
```

```python
cuisine_type = zomato_df['Cuisine'].str.contains('Biryani|Chettinad|Seafood|Hyderabadi|Kerala|Andhra|Tamil', case=False, na=False).sum()
print(f"Total restaurants offering South Indian Food: {cuisine_type}")
cuisine_type = zomato_df['Cuisine'].str.contains('Hot Chocolate|Pancakes|Sandwich|Burgers|Cheesecake|Healthy Food|Juices|Burger|Fast Food|Finger Food|Beverages|Cafe|Steak|Salad|Desserts|Street Food|Mithai|Pizza|Bakery', case=False, na=False).sum()
print(f"Total restaurants offering Cafe/Bakery Food: {cuisine_type}")
cuisine_type = zomato_df['Cuisine'].str.contains('North Indian|Mughlai|Rolls|Rajasthani|Bakery', case=False, na=False).sum()
print(f"Total restaurants offering North Indian Food: {cuisine_type}")
cuisine_type = zomato_df['Cuisine'].str.contains('Momos|Burmese|Thai|Vietnamese|Indonesian|Japanese|Singaporean|European|Middle Eastern|Chinese|American|BBQ|Mexican|Arabian|Lebanese|Italian|Continental|Asian|Kebab|Malaysian', case=False, na=False).sum()
print(f"Total restaurants offering Continental Food: {cuisine_type}")
```

```python
south_indian = 'Biryani|Chettinad|Seafood|Hyderabadi|Kerala|Andhra|Tamil'
north_indian = 'North Indian|Mughlai|Rolls|Rajasthani'
cafe_bakery = 'Hot Chocolate|Pancakes|Sandwich|Burgers|Cheesecake|Healthy Food|Juices|Burger|Fast Food|Finger Food|Beverages|Cafe|Steak|Salad|Desserts|Street Food|Mithai|Pizza|Bakery'
continental = 'Momos|Burmese|Thai|Vietnamese|Indonesian|Japanese|Singaporean|European|Middle Eastern|Chinese|American|BBQ|Mexican|Arabian|Lebanese|Italian|Continental|Asian|Kebab|Malaysian'
counts = {'South Indian Food': zomato_df['Cuisine'].str.contains(south_indian, case=False, na=False).sum(),
    'Cafe/Bakery Food': zomato_df['Cuisine'].str.contains(cafe_bakery, case=False, na=False).sum(),
    'North Indian Food': zomato_df['Cuisine'].str.contains(north_indian, case=False, na=False).sum(),
    'Continental Food': zomato_df['Cuisine'].str.contains(continental, case=False, na=False).sum()}
cuisine_type_df = pd.DataFrame(list(counts.items()), columns=['Cuisine Type', 'Restaurant Count'])
print(cuisine_type_df)
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x='Cuisine Type', y='Restaurant Count', data=cuisine_type_df)
plt.title('COUNT OF RESTAURANTS BASED ON TYPE OF CUISINE AVAILABLE ')
plt.xlabel('Cuisine Type')
plt.ylabel('Restaurant Count')
```
