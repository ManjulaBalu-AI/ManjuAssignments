```python
import pandas as pd
```

```python
dataset=pd.read_csv("Travel.csv")
```

```python
dataset
```

```python
dataset["NumberOfTrips"].fillna(0,inplace=True)
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
```

```python
dataset["DurationOfPitch"].fillna(dataset["DurationOfPitch"].median(),inplace=True)
```

```python
dataset["NumberOfFollowups"].fillna(dataset["NumberOfFollowups"].median(),inplace=True)
```

```python
dataset["PreferredPropertyStar"].fillna(dataset["PreferredPropertyStar"].mean(),inplace=True)
```

```python
dataset["NumberOfChildrenVisiting"].fillna(dataset["NumberOfChildrenVisiting"].mean(),inplace=True)
```

```python
dataset["MonthlyIncome"].fillna(0,inplace=True)
dataset["TypeofContact"].fillna('Nil',inplace=True)
```

```python
dataset.isna().sum()
```

```python
from scipy.stats import ttest_ind
income=dataset[dataset["MonthlyIncome"]>20000]["NumberOfTrips"]
male=dataset[dataset["Gender"]=='Male']["NumberOfTrips"]
ttest_ind(income,male)
```

```python
#here the t-test value or p-value is 200% which is not a acceptable one
```

```python
#if the P-value>0.05 then Null hypothesis is rejected and Alternative hypothesis get accepted
# so, here the Alternative hypothesis is aceepted and Null hypothesis is rejected 
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.DataFrame(dataset)
```

```python
model = ols('NumberOfTrips ~ C(MonthlyIncome) + C(Age) + C(MonthlyIncome):C(Age)', data=dataset).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
